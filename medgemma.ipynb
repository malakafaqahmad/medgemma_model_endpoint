```python
from fastapi import FastAPI, UploadFile, File, Form
from fastapi.responses import JSONResponse
from fastapi.middleware.cors import CORSMiddleware
from transformers import pipeline
from PIL import Image
import requests
import torch
import nest_asyncio
import uvicorn
import asyncio
import uuid
import json
import io

nest_asyncio.apply()

# --- Initialize pipeline ---
pipe = pipeline(
    "image-text-to-text",
    model="google/medgemma-4b-it",
    torch_dtype=torch.bfloat16,
    device="cuda",
)

# --- FastAPI ---
app = FastAPI(title="MedGemma Multimodal Chat API")
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_methods=["*"],
    allow_headers=["*"],
)

@app.get("/")
async def root():
    return {"status": "Server is running 🚀"}

# --- Helper functions ---
def load_image_from_url(url: str) -> Image.Image:
    """Load image from URL and return PIL Image"""
    resp = requests.get(url, headers={"User-Agent": "medgemma-api"}, stream=True)
    resp.raise_for_status()
    return Image.open(io.BytesIO(resp.content)).convert("RGB")

def load_image_from_upload(upload_file: UploadFile) -> Image.Image:
    """Load image from UploadFile and return PIL Image"""
    upload_file.file.seek(0)
    return Image.open(upload_file.file).convert("RGB")

def format_messages(raw_messages: list, files: list[UploadFile]) -> list:
    """Convert raw client messages to format expected by pipeline"""
    file_map = {f.filename: f for f in files}
    formatted = []

    for msg in raw_messages:
        new_msg = {"role": msg["role"], "content": []}
        for c in msg.get("content", []):
            if c.get("type") == "image":
                img = None
                if "image_url" in c:
                    try:
                        img = load_image_from_url(c["image_url"])
                    except Exception as e:
                        new_msg["content"].append({"type": "error", "text": f"Failed to load URL image: {e}"})
                        continue
                elif "image_file" in c:
                    filename = c["image_file"]
                    if filename in file_map:
                        try:
                            img = load_image_from_upload(file_map[filename])
                        except Exception as e:
                            new_msg["content"].append({"type": "error", "text": f"Failed to load file image: {e}"})
                            continue
                if img:
                    new_msg["content"].append({"type": "image", "image": img})
            else:
                new_msg["content"].append(c)
        formatted.append(new_msg)
    return formatted

# --- Stateless endpoint ---
@app.post("/respond")
async def respond(
    messages: str = Form(...),
    files: list[UploadFile] = File(None),
):
    messages = json.loads(messages)
    files = files or []

    formatted_messages = format_messages(messages, files)

    print("what the model sees [respond]:", formatted_messages)

    output = pipe(text=formatted_messages, max_new_tokens=3000)
    assistant_reply = output[0]["generated_text"][-1]["content"]
    print("Assistant reply:", assistant_reply)

    return JSONResponse({"response": assistant_reply})

# --- Stateful chat endpoint ---
CHAT_STORE = {}

@app.post("/chat")
async def chat(
    messages: str = Form(...),
    conversation_id: str = Form(None),
    files: list[UploadFile] = File(None),
):
    messages = json.loads(messages)
    files = files or []

    if not conversation_id:
        conversation_id = str(uuid.uuid4())
        CHAT_STORE[conversation_id] = []

    history = CHAT_STORE.get(conversation_id, [])

    formatted_messages = format_messages(messages, files)
    history.extend(formatted_messages)

    print(f"processing query for conversation {conversation_id} ...")
    output = pipe(text=history, max_new_tokens=3000)
    assistant_reply = output[0]["generated_text"][-1]["content"]

    # Add assistant reply to history
    history.append({"role": "assistant", "content": [{"type": "text", "text": assistant_reply}]})
    CHAT_STORE[conversation_id] = history

    return JSONResponse({"conversation_id": conversation_id, "response": assistant_reply})

# --- Run server ---
if __name__ == "__main__":
    print("Server starting at http://localhost:8000")
    config = uvicorn.Config(app=app, host="0.0.0.0", port=8000, log_level="info")
    server = uvicorn.Server(config=config)
    asyncio.get_event_loop().run_until_complete(server.serve())
```
